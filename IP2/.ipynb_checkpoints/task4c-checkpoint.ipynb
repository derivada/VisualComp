```python
import skimage
import skimage.io
import skimage.transform
import os
import numpy as np
import utils
import matplotlib.pyplot as plt
```

```python
if __name__ == "__main__":
    # DO NOT CHANGE
    impath = os.path.join("images", "noisy_moon.png")
    im = utils.read_im(impath)

    ### START YOUR CODE HERE ### (You can change anything inside this block)

    im_fft = np.fft.fft2(im)

    filter = np.ones(im.shape)
    rr, cc = skimage.draw.ellipse(269, 145, 5, 5)
    filter[rr, cc] = 0
    rr, cc = skimage.draw.ellipse(269, 175, 5, 5)
    filter[rr, cc] = 0
    rr, cc = skimage.draw.ellipse(269, 203, 5, 5)
    filter[rr, cc] = 0
    rr, cc = skimage.draw.ellipse(269, 232, 5, 5)
    filter[rr, cc] = 0
    rr, cc = skimage.draw.ellipse(269, 261, 5, 5)
    filter[rr, cc] = 0
    rr, cc = skimage.draw.ellipse(269, 290, 5, 5)
    filter[rr, cc] = 0
    rr, cc = skimage.draw.ellipse(269, 319, 5, 5)
    filter[rr, cc] = 0
    
    filter = np.fft.fftshift(filter)
    conv_fft = im_fft * filter
    
    im_filtered = np.fft.ifft2(conv_fft).real

    # Use plt.subplot to place two or more images beside eachother
    plt.figure(figsize=(20, 4))
    # plt.subplot(num_rows, num_cols, position (1-indexed))
    plt.subplot(1, 5, 1)
    plt.imshow(im, cmap="gray")
    # Visualize filtered spatial image
    plt.subplot(1, 5, 2) 
    plt.imshow(np.log(np.absolute(np.fft.fftshift(im_fft)) + 1))
    plt.title("Original image freq. domain")
    plt.subplot(1, 5, 3) 
    plt.imshow(np.fft.fftshift(filter))
    plt.title("Filter")
    plt.subplot(1, 5, 4) 
    plt.imshow(np.log(np.absolute(np.fft.fftshift(conv_fft)) + 1))
    plt.title("Filtered image freq. domain")
    plt.subplot(1, 5, 5) 
    plt.imshow(im_filtered, cmap="gray")
    plt.show()
    ### END YOUR CODE HERE ###
    utils.save_im("moon_filtered.png", utils.normalize(im_filtered))
    utils.save_im("moon_filtered.png", utils.normalize(np.log(np.absolute(np.fft.fftshift(conv_fft)) + 1)))
```
